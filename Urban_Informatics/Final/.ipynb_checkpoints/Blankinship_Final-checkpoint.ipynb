### 4.2 Calculate the fraction of asthma and drug related diagnoses per facility
* count the number of asthma and drug related diagnoses per facility

Find in the file all asthma-related diagnosis : I used those that include asthma and respiratory issues as diagnosis that are relevant, and I used alcohol and drug related issues as well as poisoning from psychotropic substances as relevant drug-related diagnoses hint

* calculate the fraction of asthma/drug diagnoses over all diagnoses per facility

```python
# Here are my masks
masks = []
keywords = ['Asthma','Respiratory','pyschotropic','Substance','Alcohol','Poisoning by other medications']

for k in keywords:
    masks.append(SPARCS['CCS Diagnosis Description'].str.contains(k))
```

```python
mask = [any(x) for x in zip(masks[0], masks[1])]
mask2 = [any(x) for x in zip(masks[2], masks[3], masks[4], masks[5])]
```

```python
asthma = SPARCS[mask]
drugs = SPARCS[mask2]
```

```python
asthma.shape, drugs.shape
```

```python
countAll = SPARCS[['Facility ID','CCS Diagnosis Code']].groupby('Facility ID').count()
countAll.rename(columns={'CCS Diagnosis Code': 'NumPatients'}, inplace=True)
countAll.head()
```

```python
countAsthma = asthma[['Facility ID','CCS Diagnosis Code']].groupby('Facility ID').count()
countAsthma.rename(columns={'CCS Diagnosis Code': 'AsthmaCount'}, inplace=True)
countAsthma.head()
```

```python
countDrugs = drugs[['Facility ID','CCS Diagnosis Code']].groupby('Facility ID').count()
countDrugs.rename(columns={'CCS Diagnosis Code': 'DrugCount'}, inplace=True)
countDrugs.head()
```

```python
facilities = countAll.merge(countAsthma, left_index=True, right_index=True)
facilities = facilities.merge(countDrugs, left_index=True, right_index=True)

facilities['asthmaRatio'] = facilities.AsthmaCount/facilities.NumPatients
facilities['drugsRatio'] = facilities.DrugCount/facilities.NumPatients

facilities.head()
```

```python
print("Number of NYC discharges: ", np.sum(facilities.NumPatients))
print("number of NYC facilities in the dataset", len(facilities))
print("number of asthma related cases in NYC", np.sum(facilities.AsthmaCount)) 
print("number of drug related cases in NYC", np.sum(facilities.DrugCount)) 
```

### 6.2 Plot the SAT feature by PUMA against 3 of the crime features

```python
# models won't work if there are nan values in one category but not another
SAT_Crime = alldata.dropna(subset=['averageScore'])
SAT_Crime.shape
```

```python
plt.figure(figsize=(15,5))
plt.suptitle("Average SAT Score per PUMA against Crimes",fontsize=25)
y = SAT_Crime.averageScore
crimes = ['AvgOfMajor N', 'AvgOfOth N', 'AvgOfNoCrim N']
s = 130 

for i in range(3):
    s+=1
    avgcrim_const = sm.add_constant(SAT_Crime[crimes[i]])
    mod = smf.OLS(y, avgcrim_const).fit()
    plt.subplot(s)
    plt.scatter(SAT_Crime[crimes[i]], y)
    plt.plot(SAT_Crime[crimes[i]],mod.predict(avgcrim_const), 
         'k*-', label='R2: {0:.3f}'.format(mod.rsquared))
    plt.xlabel("Average Crime Factor: {}".format(crimes[i]))
    plt.ylabel("Average SAT score")
    plt.legend()
```

**Figure 8:** All of the plots above appear very similar but we can see by the corresponding $R^2$ values that the **Average of Major Crimes** had the best fit against Average SAT scores.

create, the models, evaluate their performance, assess the feature importance.

```python
# Had to model my own since PUI kernel doesnt have sklearn.model_selection
def train_test_split(x,y,test_size=0.33):
    '''This replicates the train test split of sci kit learn'''
    mask = (np.random.rand(len(x)) > test_size)
    return(x[mask],x[~mask],y[mask],y[~mask])    
```

### 10.1 create a random forest model using only the crime in school data, but using all 10 crime in school features (5 points)
measure the model accuracy, precision, recall and the feature importance

```python
crimes = ['Major N', 'Oth N', 'NoCrim N', 'Prop N', 'Vio N',
       'AvgOfMajor N', 'AvgOfOth N', 'AvgOfNoCrim N', 'AvgOfProp N',
       'AvgOfVio N']
```

```python
rfData = alldata.dropna(subset=['averageScore'])
```

Here I am standardizing data and treating each of the above columns as independent

```python
stdValues=crimes.copy()
stdValues.append('averageScore')
rfData[stdValues]=(rfData[stdValues]-rfData[stdValues].mean())/rfData[stdValues].std()
```

```python
rfData[stdValues].describe()
```

OK, the features we are concerned with all have a mean of 0 and standard deviation of 1. Next we want to set up cross validation for the model to account for the train and test splits. This is useful since we have a small amount of sample data and it ensures that observations will be included in both the training and test sets.

```python
def CrossValidation(n, df, x_variables, splitRatio=0.33):
    '''
    This function takes runs cross validation for our Random Forest Regression.
    
    Input
    --------------
    n: number of time to run
    df: dataframe (so I can run this again later)
    splitRatio (optional): what the train,test split should be
    
    
    Return
    --------------
    (Average R^2 Score, Average Train score, Average Test Score)
    '''
    np.random.seed(0)
    # Apply cross-validation to diagnose model for overfitting  
    TrainScore=[] #training R2 Score
    TestScore=[] #out-of-sample R2 score
    
    for i in range(n):
        X_train, X_test, y_train, y_test = train_test_split(df[x_variables], 
                                                            df['averageScore'],
                                                           test_size=splitRatio)
        RFR = rfr().fit(X_train,y_train) 
        y_pred = RFR.predict(X_test)
        TrainScore.append(RFR.score(X_train,y_train))
        TestScore.append(RFR.score(X_test,y_test))
               
    return(np.mean(TrainScore), np.mean(TestScore))
```

```python
#Cross Validating 10 times
CrossValidation(100, rfData, crimes, 0.33)
```

**Interpretation**: We can see that our model is overfitting. The $R^2$ of our in-sample data is much, much higher than our out of sample $R^2$ statistic, and this is backed up by cross validation.
